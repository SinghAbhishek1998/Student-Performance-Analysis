```python
Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
IQR = Q3 - Q1    #IQR is interquartile range. 

filter = (df['math score'] >= Q1 - 1.5 * IQR) & (df['math score'] <= Q3 + 1.5 *IQR)
df.loc[filter] 
```

```python
Q1 = df['reading score'].quantile(0.25)
Q3 = df['reading score'].quantile(0.75)
IQR = Q3 - Q1    #IQR is interquartile range. 

filter = (df['reading score'] >= Q1 - 1.5 * IQR) & (df['reading score'] <= Q3 + 1.5 *IQR)
df.loc[filter] 
```

```python
Q1 = df['writing score'].quantile(0.25)
Q3 = df['writing score'].quantile(0.75)
IQR = Q3 - Q1    #IQR is interquartile range. 

filter = (df['writing score'] >= Q1 - 1.5 * IQR) & (df['writing score'] <= Q3 + 1.5 *IQR)
df.loc[filter] 
```
